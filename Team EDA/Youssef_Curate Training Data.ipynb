```python

import os
import pandas as pd
curr_path: str = os.getcwd() 
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 10000)
```

```python
curr_path
```

```python

# Define the screw types
training_data = {'file_path':[],'keywords':[]}

# Path to the directory containing the images
image_directory = os.path.dirname(curr_path) + '\\data'

# load product attr table
df = pd.read_csv(os.path.dirname(curr_path) + r'\\screw_category-name_product-attribute(in).csv')
df.attributename = df.attributename.apply(lambda x: x.lower())


# parses the words that matter within each file name
for filename in os.listdir(image_directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        keywords = filename.split('-')[7:-2]
        training_data['file_path'].append(filename)
        training_data['keywords'].append(' '.join(keywords))





        
image_file_keywords_df = pd.DataFrame(training_data)
        

```

```python
' '.join(keywords)
```

```python
pd.set_option('max_colwidth', 800)
image_file_keywords_df.head(50)
```

```python
for idx in image_file_keywords_df.index:
    image_file_word: str = image_file_keywords_df.loc[idx,'keywords']
    
    # search for the keyword in the file name in the product list and match it all up
    idx_of_interest: list[str] = df[df['attributename'].str.contains(image_file_word)].index.tolist()

    if len(idx_of_interest) != 0:

        df_queried: list = df.iloc[idx_of_interest][['category_name','attributename']].drop_duplicates().values.tolist()[0]
        capture_file_name = image_file_keywords_df.loc[idx,'file_path']
        ix_to_replace = image_file_keywords_df[image_file_keywords_df['file_path'] == capture_file_name].index.tolist()
        
        for word in df_queried:
            image_file_keywords_df.loc[ix_to_replace,'extra_names'] = word
```
